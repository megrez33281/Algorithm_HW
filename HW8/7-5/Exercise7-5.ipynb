```python
#Exercise7-5  4 () = 7^ (mod 15) 
#N=15 a=7

from qiskit import QuantumRegister, QuantumCircuit, ClassicalRegister,execute
from qiskit.providers.aer import AerSimulator
from qiskit.visualization import plot_histogram
from math import pi
from fractions import Fraction

def qc_mod15(a,power):
    #amod 15U
    assert a in [2,4,7,8,11,13], 'Invalid value of argument a:'+str(a)
    qrt = QuantumRegister(4,'target') #0~1424
    U = QuantumCircuit(qrt)
    for i in range(power):
        if a in [2,13]:
            U.swap(0,1)
            U.swap(1,2)
            U.swap(2,3)
        if a in [7,8]:
            U.swap(2,3)
            U.swap(1,2)
            U.swap(0,1)
        if a in [4,11]:
            U.swap(1,3)
            U.swap(0,2)
        if a in[7,11,13]:
            for j in range(4):
                U.x(j)
    U = U.to_gate()
    U.name = f'{a}^{power}mod 15'
    C_U = U.control() #
    return C_U

def iqft(n):
    #
    br = QuantumRegister(n,'b')
    qc = QuantumCircuit(br)
    for sbit in range(n//2):
        qc.swap(sbit,n-sbit-1)#swap
    for hbit in range(0,n,1):
        for cbit in range(hbit-1,-1,-1):
            qc.cp(-pi/2**(hbit-cbit),cbit,hbit)#control p
        qc.h(hbit)#h
    return qc

def qpf15(count_number,a):
    target_register = QuantumRegister(target_number,'target')
    y_register = QuantumRegister(4,'y')#mod 15
    creg = ClassicalRegister(target_number,'c')
    qc = QuantumCircuit(target_register,y_register,creg)
    for hbit in range(count_number):
        qc.h(hbit)
    qc.x(y_register[0]) #|1>
    #U
    for cbit in range(count_number):
        qc.append(qc_mod15(a,2**cbit),[cbit] + list(range(count_number,count_number+4)))#cbitcount_number~count_number+3(y_register)
    qc.append(iqft(count_number).to_gate(label='IQFT'),range(count_number)) #target
    qc.measure(range(count_number),creg)
    return qc

target_number  = 4 #
a = 7
qc = qpf15(target_number,a)
display(qc.draw('mpl'))

#
sim = AerSimulator()
job = execute(qc,backend=sim,shots=1000)
result = job.result()
counts = result.get_counts(qc)
display(plot_histogram(counts))
print("Total counts for qubit states are:",counts)
print('%10s %10s %10s %10s %10s' % ('Binary','Decimal','Phase','Fraction','Peroid'))
for akey in counts.keys():
    dec = int(akey,base=2)#
    phase = dec/(2**target_number)#eigenvalue = e^(2*pi*i*phase)i
    frac=Fraction(phase).limit_denominator(15) #15peroid
    period=frac.denominator #phase = s/rrperoid
    print('%10s %10s %10s %10s %10s' % (akey,dec,phase,frac,period))
```
